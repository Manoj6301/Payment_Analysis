```python
import pandas as pd
from google.colab import files
import matplotlib.pyplot as plt
import seaborn as sns



# Load datasets into pandas DataFrames
account_df = pd.read_csv("sample_data/Account Info.csv")  # Replace with your actual file name
payment_df = pd.read_csv("sample_data/Payment Info.csv")

# Display the first few rows
print("Account Info Data:")
print(account_df.head())

print("\nPayment Info Data:")
print(payment_df.head())
```

Data Exploration & Cleaning

```python
# Check basic info
print(account_df.info())
print(payment_df.info())

# Check for missing values
print(account_df.isnull().sum())
print(payment_df.isnull().sum())

# Convert dates to datetime format
account_df["Policy Start Date"] = pd.to_datetime(account_df["Policy Start Date"])
account_df["Policy Expiry Date"] = pd.to_datetime(account_df["Policy Expiry Date"])
payment_df["Payment Due Date"] = pd.to_datetime(payment_df["Payment Due Date"])
```

The bar chart illustrates that General Liability policies significantly outnumber Auto policies. Identifying Late Payments by Region

```python
late_payments = payment_df[payment_df["Payment Status"] == "Late"]
merged_data = pd.merge(late_payments, account_df, on="Policyholder ID", how="left")

plt.figure(figsize=(8,5))
sns.countplot(data=merged_data, x="Region", palette="coolwarm")
plt.title("Late Payments by Region")
plt.xticks(rotation=45)
plt.show()
```

The bar chart clearly highlights Texas as having the highest number of late payments, followed by states such as Florida, Illinois, and California. This indicates significant regional variation in payment behavior and suggests targeted measures, such as region-specific payment reminders or incentives, could effectively improve payment compliance, particularly in Texas and other top regions experiencing delays

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Merge account and payment data
merged_df = pd.merge(payment_df, account_df, on="Policyholder ID", how="left")

# Group by Policy Type and Payment Method
policy_payment_counts = merged_df.groupby(["Policy Type", "Payment Method"]).size().unstack()

# Plot the payment method distribution for each policy type
plt.figure(figsize=(8,5))
policy_payment_counts.plot(kind="bar", stacked=True, colormap="viridis", figsize=(10,6))
plt.title("Payment Methods Used for Different Policy Types")
plt.ylabel("Count of Payments")
plt.xticks(rotation=45)
plt.legend(title="Payment Method")
plt.show()
```

The visualization clearly shows the preference of payment methods by policy type. For General Liability policies, ACH payments are predominantly utilized, followed by Credit Cards and Checks. Conversely, for Auto policies, payments via Credit Card are most common, followed by Checks, and ACH methods being the least utilized. This suggests that policyholders might find ACH convenient for larger or recurring payments typical in general liability scenarios, whereas auto policyholders may prefer immediate payments using credit cards. Emphasizing ACH for auto policies could increase efficiency and reliability in payment collection.

```python
# Count the number of policyholders per state (region)
policyholder_counts = account_df.groupby("Region")["Policyholder ID"].nunique().reset_index()

# Sort by highest number of policyholders
policyholder_counts = policyholder_counts.sort_values(by="Policyholder ID", ascending=False)

# Display the state with the highest policyholders
top_state = policyholder_counts.iloc[0]
print(f"State with the highest volume of policyholders: {top_state['Region']} ({top_state['Policyholder ID']} policyholders)")

# Plot the number of policyholders per state
plt.figure(figsize=(10,6))
sns.barplot(data=policyholder_counts, x="Region", y="Policyholder ID", palette="viridis")
plt.xticks(rotation=45)
plt.title("Number of Policyholders by State")
plt.xlabel("State (Region)")
plt.ylabel("Number of Policyholders")
plt.show()
```

The bar chart shows that California has the highest number of policyholders, significantly surpassing other states. New York and Texas follow as the second and third highest, respectively. The concentration in these populous states highlights strategic regions for targeted marketing, customer engagement, and service improvement initiatives. Given this distribution, it’s advisable to prioritize customer service enhancements and payment compliance measures specifically in these high-volume states

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count the occurrences of each payment status
payment_status_counts = payment_df["Payment Status"].value_counts()

# Calculate percentages
payment_status_percent = (payment_status_counts / payment_status_counts.sum()) * 100

# Display results
print("Payment Status Breakdown:")
print(payment_status_percent)

# Plot the distribution using a pie chart
plt.figure(figsize=(6,6))
payment_status_percent.plot(kind="pie", autopct="%.1f%%", colors=["green", "red"], startangle=90, wedgeprops={"edgecolor": "black"})
plt.title("Percentage of On-Time vs. Late Payments")
plt.ylabel("")  # Hide y-label
plt.show()
```

The pie chart distinctly shows that a majority (56%) of payments were late, compared to 44% that were on time. This considerable proportion of late payments indicates significant inefficiencies or barriers in the current payment collection system. To address this, the company could benefit from enhanced reminder systems, automated payment processing options, or incentives for timely payments

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count occurrences of each Payment Source
payment_source_counts = payment_df["Payment Source"].value_counts()

# Calculate percentages
payment_source_percent = (payment_source_counts / payment_source_counts.sum()) * 100

# Display results
print("Payment Source Breakdown:")
print(payment_source_percent)

# Plot the distribution using a pie chart
plt.figure(figsize=(6,6))
payment_source_percent.plot(kind="pie", autopct="%.1f%%", colors=["blue", "orange"], startangle=90, wedgeprops={"edgecolor": "black"})
plt.title("Percentage of Payments: Policyholders vs. Agents")
plt.ylabel("")  # Hide y-label
plt.show()
```

The pie chart clearly depicts that 58% of payments are made directly by Policyholders, while Agents account for 42% of payments. This suggests policyholders predominantly manage their payments directly, but agent-based payments also constitute a substantial portion. Efforts to streamline payment processing should consider enhancing user-friendly direct payment methods for policyholders, while also improving agent-assisted payment systems.

```python
payment_df["Total Outstanding Amount"] = payment_df["Total Outstanding Amount"].replace('[\$,]', '', regex=True).astype(float)
payment_df["Payment Amount"] = payment_df["Payment Amount"].replace('[\$,]', '', regex=True).astype(float)

# Display updated data
print(payment_df[["Total Outstanding Amount", "Payment Amount"]].head())
```

```python
import pandas as pd

# Ensure no division by zero errors
payment_df = payment_df.copy()
payment_df["Payment Amount"] = payment_df["Payment Amount"].replace(0, float("nan"))

# Calculate remaining payments
payment_df["Remaining Payments"] = (payment_df["Total Outstanding Amount"] / payment_df["Payment Amount"]).fillna(0).round(0)

# Display relevant columns
print(payment_df[["Policyholder ID", "Total Outstanding Amount", "Payment Amount", "Remaining Payments"]].head())

# Find policyholders with the highest remaining payments
top_outstanding = payment_df.sort_values(by="Remaining Payments", ascending=False)
print("\nTop Policyholders with the Most Remaining Payments:")
print(top_outstanding[["Policyholder ID", "Remaining Payments"]].head(10))
```
